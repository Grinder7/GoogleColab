```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-poster')
x = [0, 1, 2]
y = [1, 3, 2]
```

```python
f = interp1d(x, y)
y_hat = f(1.5)
print(y_hat)
```

```python
plt.figure(figsize = (10, 8))
plt.plot(x, y, '-ob')
plt.plot(1.5, y_hat, 'ro')
plt.title('Linear Interpolation at x = 1.5')
plt.xlabel('x')
plt.xlabel('y')
plt.show()
```
